imdb score > 8, content_rating = pg-13, title year <2000 or >2009

```python
imdb_score = movie.imdb_score > 8
content_rating = movie.content_rating == 'PG-13'
title_year = (movie.title_year < 2000) | (movie.title_year > 2009)
```

```python
overall_result = imdb_score & content_rating & title_year

overall_result.head()
```

```python
imdb_score2 = movie.imdb_score < 5
content_rating2 = movie.content_rating == 'R'
title_year2 = movie.title_year.between(2000, 2010)
```

```python
overall_result2 = imdb_score2 & content_rating2 & title_year2
overall_result2
```

```python
final_overall = overall_result | overall_result2

final_overall
```

```python
movie[final_overall].head()
```

```python
movie.loc[final_overall]
```

```python
movie.loc[final_overall]
```

```python
cols = ['imdb_score', 'num_critic_for_reviews', 'duration']

movie_filtered = movie.loc[final_overall, cols]
movie_filtered
```

```python
import pandas as pd
employee = pd.read_csv('employee.csv')
employee
```

```python
dep = [
    'Houston Police Department-HPD',
    'Houston Fire Department (HFD)'
]

dept_new = employee.DEPARTMENT.isin(dep)
gender = employee.GENDER == 'Female'
base_salary = employee.BASE_SALARY.between(80_000, 120_000)
```

```python
final_result = dept_new & gender & base_salary
final_result
```

```python
select_columns = [
    'UNIQUE_ID',
    'DEPARTMENT',
    'GENDER',
    'BASE_SALARY'
]
```

```python
employee.loc[final_result, select_columns]
```

```python
SELECT *
FROM EMPLOYEE
WHERE DEPARTMENT not in (
    SELECT DEPARTMENT
    FROM (SELECT TOP 5 DEPARTMENT,
                 COUNT(1) as CT
          FROM EMPLOYEE
          GROUP BY DEPARTMENT
          ORDER BY CT DESC
          
          ) 
)
```
