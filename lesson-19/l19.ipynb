# concat

```python
stocks_2016 = pd.read_csv('stocks_2016.csv')
stocks_2017 = pd.read_csv('stocks_2017.csv')
```

```python
stocks_2016
```

```python
stocks_2017
```

```python
ls = [stocks_2016, stocks_2017]

pd.concat(ls)
```

```python
pd.concat(ls, keys=['2016', '2017'], names=['Year', 'Symbol'])
```

```python
pd.concat(ls, keys=['2016', '2017'], axis='columns', names=['Year', None])
```

```python
pd.concat(ls, join='inner', keys=['2016', '2017'], axis=1, names=['Year', None])
```

# Join and Merge

```python
years = 2016, 2017, 2018

stocks_years = [pd.read_csv(f'stocks_{year}.csv', index_col='Symbol') for year in years]
```

```python
stocks_2016, stocks_2017, stocks_2018 = stocks_years
```

```python
stocks_2018
```

```python
stocks_2017
```

```python
stocks_2016.join(stocks_2017, on= 'Symbol', lsuffix='_2016', rsuffix='_2017', how= 'inner')
```

```python
stocks_2016.join(stocks_2017, lsuffix='_2016', rsuffix='_2017', how= 'left')
```

```python
stocks_2016.join(stocks_2017, lsuffix='_2016', rsuffix='_2017', how= 'cross')
```

```python
stocks_2016.join(stocks_2017, lsuffix='_2016', rsuffix='_2017', how= 'outer')
```

```python
other = [
    stocks_2017.add_suffix('_2017'),
    stocks_2018.add_suffix('_2018')
]

#stocks_2016.add_suffix('_2016').join(stocks_2017.add_suffix(('_2017')), how='outer').join(stocks_2018.add_suffix('_2018'))
stocks_2016.add_suffix('_2016').join(other, how= 'outer')
```

```python
stocks_2016.merge(stocks_2017, left_index=True, right_index=True, how='inner')
```

```python
df1 = pd.DataFrame({'Value_A': [10, 20, 30]}, index=['a', 'b', 'c'])
df2 = pd.DataFrame({'Value_B': [100, 200, 300]}, index=['b', 'c', 'd'])

merged = df1.merge(df2, left_index=True, right_index=True)
merged
```

```python
stocks_2016.merge(stocks_2017, 
                  left_index=True, 
                  right_index=True, 
                  how='outer', 
                  suffixes=('_2016', '_2017')
                  ).merge(stocks_2018.add_suffix('_2018'), 
                          left_index=True, 
                          right_index=True, 
                          how='outer')
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite:///data/chinook.db')
```

```python
tracks = pd.read_sql_table('tracks', engine)

tracks
```

```python
genres = pd.read_sql_table('genres', engine)
genres
```

```python
genres.merge(tracks[['Milliseconds', 'GenreId']], on='GenreId', how='right').drop('GenreId', axis='columns')
```

```python
cust = pd.read_sql_table('customers', engine, columns=['CustomerId', 'FirstName', 'LastName'])
invoice = pd.read_sql_table('invoices', engine, columns=['InvoiceId', 'CustomerId'])
invoice_items = pd.read_sql_table('invoice_items', engine, columns=['InvoiceId', 'UnitPrice', 'Quantity'])
```

```python
cust.merge(invoice, on= 'CustomerId', how='inner').merge(invoice_items, on='InvoiceId')
```

```python
sql_string = '''
    select * from invoices where CustomerId = 3
'''

pd.read_sql_query(sql_string, engine)
```

```python
sql_string1 = '''
SELECT Name,
    time(avg(Milliseconds) / 1000, 'unixepoch') as avg_time
FROM (
    SELECT
        g.Name,
        t.Milliseconds
    FROM genres as g
    JOIN tracks as t on g.genreid == t.genreid
)
GROUP BY Name
ORDER BY avg_time'''

pd.read_sql_query(sql_string1, engine)
```
