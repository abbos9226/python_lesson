1.Group the data by the Category column and calculate the following aggregate statistics for each category:
    Total quantity sold.
    Average price per unit.
    Maximum quantity sold in a single transaction.
2.Identify the top-selling product in each category based on the total quantity sold.
3.Find the date on which the highest total sales (quantity * price) occurred.

```python
import pandas as pd
```

```python
df=pd.read_csv('sales_data.csv')
df
```

```python
#1.Group the data by the Category column and calculate the following aggregate statistics for each category:
    #Total quantity sold.
    #Average price per unit.
    
av_price=df.groupby('Category').agg(
    total_sold=('Quantity','sum'),
    av_price_per=('Price','mean')
)

av_price
```

```python
#Maximum quantity sold in a single transaction.
category_stats=df.groupby('Category').agg(per=('Quantity','max'))
category_stats 
```

```python
df.head(2)
```

```python
#2.Identify the top-selling product in each category based on the total quantity sold.

# Group by 'Category' and 'Product', then sum the quantities
product_sales = df.groupby(['Category', 'Product'])['Quantity'].sum().reset_index()

# Sort by Category and Quantity (descending) to find top products
product_sales_sorted = product_sales.sort_values(['Category', 'Quantity'], ascending=[True, False])

# Get the top-selling product in each category
top_selling_products = product_sales_sorted.groupby('Category').first().reset_index()

# Rename column for clarity
top_selling_products.columns = ['Category', 'Top_Product', 'Total_Quantity_Sold']

top_selling_products 
```

```python
df.head(2)
```

```python
#3.Find the date on which the highest total sales (quantity * price) occurred.

# Calculate total sales per row
df['Sales'] = df['Quantity'] * df['Price'] 

# Group by Date and sum the Sales
daily_sales = df.groupby('Date')['Sales'].sum().reset_index()

# Find the date with the highest sales
max_sales_date = daily_sales.loc[daily_sales['Sales'].idxmax()]

max_sales_date
```

Homework Assignment 3: Population Salary Analysis

1."task\population.db" sqlite database has population table.
2."task\population salary analysis.xlsx" file defines Salary Band categories.
    Read the data from population table and calculate following measures:
        Percentage of population for each salary category;
        Average salary in each salary category;
        Median salary in each salary category;
        Number of population in each salary category;
3.Calculate the same measures in each State
Note: Use SQL only to select data from database. All the other calculations should be done in python.

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('population.db')
population_df = pd.read_sql_query("SELECT * FROM population", conn)
conn.close()
```

```python
# --- Step 2: Read salary band categories from Excel file ---
salary_band_df = pd.read_excel('population salary analysis.xlsx')

# Assume salary_band_df has columns: 'Band', 'MinSalary', 'MaxSalary'
# We'll map each person into a salary band
def assign_salary_band(salary):
    row = salary_band_df[(salary_band_df['MinSalary'] <= salary) & (salary <= salary_band_df['MaxSalary'])]
    return row['Band'].values[0] if not row.empty else 'Uncategorized'

# Apply banding
population_df['SalaryBand'] = population_df['Salary'].apply(assign_salary_band)

# --- Step 3: Aggregate analysis per salary band ---
band_group = population_df.groupby('SalaryBand')

summary_by_band = band_group['Salary'].agg(
    Count='count',
    Average='mean',
    Median='median'
).reset_index()

# Calculate percentage of total population
total_count = population_df.shape[0]
summary_by_band['Percentage'] = (summary_by_band['Count'] / total_count) * 100

# --- Step 4: Aggregate analysis per State and Salary Band ---
state_band_group = population_df.groupby(['State', 'SalaryBand'])

summary_by_state_band = state_band_group['Salary'].agg(
    Count='count',
    Average='mean',
    Median='median'
).reset_index()

# Calculate percentage within each state
state_totals = population_df.groupby('State')['Salary'].count().reset_index(name='TotalInState')
summary_by_state_band = summary_by_state_band.merge(state_totals, on='State')
summary_by_state_band['Percentage'] = (summary_by_state_band['Count'] / summary_by_state_band['TotalInState']) * 100
summary_by_state_band.drop(columns='TotalInState', inplace=True)

# --- Optional: Save results to Excel or CSV ---
# summary_by_band.to_excel("salary_summary_by_band.xlsx", index=False)
# summary_by_state_band.to_excel("salary_summary_by_state_band.xlsx", index=False)

# Print preview
print("Summary by Salary Band:\n", summary_by_band.head())
print("\nSummary by State and Salary Band:\n", summary_by_state_band.head())
```
